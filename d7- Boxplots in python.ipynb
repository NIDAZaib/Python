```python
#horizontal plot

import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean 

kashti = sns.load_dataset("titanic")
kashti

# barplot
sns.barplot(x="fare", y="class", hue="sex", data = kashti, estimator=mean, saturation=0.2) # put numeric parameter on xaxis
plt.show()
```

```python
#in placeof hue add values by ownself

import seaborn as sns
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
kashti

# barplot
sns.barplot(x="class", y="fare", data = kashti,
           linewidth=2.5, facecolor=(1, 1, 1, 0), #linewidth=5 bar widht increase,errcolor=".6 " error bar color lightif 1 then 
            #then white colour it means zero is black and one is white between is grey 
           errcolor=".2", edgecolor = "0.2") #edge is border of bar is 0 to one
plt.show()
```

```python
#boxplot ()
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
# boxplot (labels)
sns.boxplot(x="survived", y="age", showmeans=True,
            meanprops= {"marker":"+", "markersize": "12", "markeredgecolor": "red"}, data = kashti)
plt.show()
#labels
plt.xlabel("how many survived", size=10, weight="bold"),
plt.ylabel("age(year)", size=10, weight="bold"),
plt.title("boxplot of kitny duby or kitny bach gye" , size=14, weight="bold") #font bold and font size increas and decrease
plt.show
```
