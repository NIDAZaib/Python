```python
#lineplot iris
#dark back
from matplotlib import rc
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style(style="dark", rc = None)
flowers = sns.load_dataset("iris")
sns.lineplot(x= "sepal_length", y= "sepal_width", data= flowers)
plt.title("Data set of iris")
sns.set_style("dark")
# plt.show()
```

```python
##size of figure
import seaborn as sns
import matplotlib.pyplot as plt

#load_dataset
flowers = sns.load_dataset("iris")
#change fig
plt.figure(figsize=(8,6)) # you can write 10,8 also and size will increase and small one too ...4,2

sns.lineplot(x= "sepal_length", y= "sepal_width", data= flowers)
plt.title("phoolon ka plot")

plt.show()
```

```python
#bar plots
import seaborn as sns
import matplotlib.pyplot as plt

#load_dataset
phool = sns.load_dataset("iris")
#change fig
plt.figure(figsize=(8,6)) # you can write 10,8 also and size will increase and small one too ...4,2

sns.barplot(x= "species", y= "sepal_width", data= phool)
plt.title("phoolon ka plot")

plt.show()
```

```python
#pelets
import seaborn as sns
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
kashti

# barplot
sns.barplot(x="sex", y="alone", hue="who", data = kashti, order = ["female", "male"], color = "red",
           palette = "pastel") #builtin pelettes for seaborn search google image  and you can use them by writing names of them.
plt.show()
```

```python
#estimator
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median

kashti = sns.load_dataset("titanic")
kashti

# barplot
sns.barplot(x="class", y="fare", hue="sex", data = kashti, estimator=median) 
plt.show()
```

```python
#estimator (mean ka)
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean ( #you can simply write import numpy it will work both for mean and median )

kashti = sns.load_dataset("titanic")
kashti

# barplot
sns.barplot(x="class", y="fare", hue="sex", data = kashti, estimator=mean) 
plt.show()
```

```python
#estimator (colour saturation) or intensity
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean 

kashti = sns.load_dataset("titanic")
kashti

# barplot
sns.barplot(x="class", y="fare", hue="sex", data = kashti, estimator=mean, saturation=0.5) #low sat do 0.1,02 high intensity 2
plt.show()
```
